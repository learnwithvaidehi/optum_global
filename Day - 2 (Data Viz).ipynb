# Data visualization (Basics)

- Representation of the data in a pictorial or graphical format
- Allow us to get the intuitive understanding of the data
- Helps to visualize the patterns in the data
- Python offers multiple great graphing libraries that come packed with lots of different features.
- Matplotlib: low level, provides lots of freedom
- Pandas Visualization: easy to use interface, built on Matplotlib
- Seaborn: high-level interface, great default styles
- Plotly: can create interactive plots
- Bokeh: used creating interactive visualizations for modern web browsers"

### Customise plots

- color, labels, title, legends, grid

```python
plt.figure(figsize=(12, 2))
plt.plot(dates, sales, color = "red", marker = "o", ls = "--")
plt.plot(dates, profits, color = "green", marker = "o", ls = "--")

plt.title("Sales over Months")
plt.grid(visible=True, axis="y", color = "orange", alpha = 0.3)

plt.legend(["Sales", "Profit"])

_ = plt.xticks(rotation = 30)
```

### Subplots

###### Visualise sales and profits as subplots

```python
fig, ax = plt.subplots(figsize=(12, 4), nrows=2, sharex=True)
ax[0].plot(dates, sales, color = "red", marker = "o", ls = "--")
ax[0].set_title("Sales over Months")
ax[0].grid(visible=True, axis="y", color = "orange", alpha = 0.3)
# ax[0].set_yticks(np.arange(0, 80000, 5000))

ax[1].plot(dates, profits, color = "green", marker = "o", ls = "--")
ax[1].set_title("Profits over Months")
ax[1].grid(visible=True, axis="y", color = "orange", alpha = 0.3)
```

### Additional chart types

```python
import numpy as np
products = np.array(['Caffe Latte', 'Cappuccino', 'Colombian', 'Darjeeling', 'Decaf Irish Cream', 'Earl Grey', 'Green Tea', 'Lemon', 'Mint', 'Regular Espresso'])
sales = np.array([52248.0, 14068.0, 71060.0, 60014.0, 69925.0, 27711.0, 19231.0, 24873.0, 32825.0, 44109.0])
profits = np.array([17444.0, 5041.0, 28390.0, 20459.0, 23432.0, 7691.0, -2954.0, 7159.0, 10722.0, 14902.0])
target_profits = np.array([15934.0, 4924.0, 31814.0, 19649.0, 24934.0, 8461.0, 7090.0, 7084.0, 10135.0, 16306.0])
target_sales = np.array([48909.0, 13070.0, 80916.0, 57368.0, 66906.0, 30402.0, 18212.0, 21628.0, 27336.0, 42102.0])
```

#### Bar chart

```python
plt.figure(figsize=(12, 2))
s_sales = np.sort(sales)[::-1]
s_products = products[np.argsort(sales)[::-1]]
plt.bar(s_products, s_sales, color = "teal", edgecolor = "black", alpha = 0.6, width = 0.6)
_ = plt.xticks(rotation = 90)
```

#### Horizantal bar chart

```python
plt.figure(figsize=(3, 4))
s_sales = np.sort(sales)[::-1]
s_products = products[np.argsort(sales)[::-1]]
plt.barh(s_products, s_sales, color = "teal", edgecolor = "black", alpha = 0.6)
_ = plt.xticks(rotation = 90)
```

#### Stacked bar chart

```python
plt.figure(figsize=(12, 4))
plt.bar(products, sales, color = "teal", edgecolor = "black", alpha = 0.6, width = 0.4)
plt.bar(products, profits, color = "cyan", edgecolor = "black", alpha = 0.6, width = 0.4)
_ = plt.xticks(rotation = 90)
_ = plt.legend(["Sales", "Profits"])
```

#### Side by Side Bar Chart

```python
plt.figure(figsize=(12, 4))
plt.bar(products, sales, color = "teal", edgecolor = "black", alpha = 0.6, width = -0.4, align = "edge")
plt.bar(products, profits, color = "cyan", edgecolor = "black", alpha = 0.6, width = 0.4, align = "edge")
plt.plot(products, sales)
# plt.axhline(np.mean(sales))
_ = plt.xticks(rotation = 90)
_ = plt.legend(["Sales", "Profits"])
```

#### Bullet chart

```python
color_list = np.where(sales > target_sales, "lightgreen", "lightcoral")

plt.figure(figsize=(4,5))
plt.barh(products, sales, color = color_list, edgecolor = "black", alpha = 0.6)
_ = plt.plot(target_sales, products, marker = "|", color = "red", ls = "", markersize = 25)
```

#### Scatter Plot

```python
plt.figure(figsize=(3, 3))
_ = plt.scatter(sales, profits)

z = np.polyfit(sales, profits, 1)
func = np.poly1d(z)
plt.plot(sales, func(sales), color = "red")
```
